# UrbanHeat: Data Analysis of Urban Heat Island Effects

This notebook demonstrates the EDA, Geospatial Analysis, and Modeling components for the UrbanHeat project.

## 1. Data Loading

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../dataset/processed_urban_heat.csv')
df.head()
```

## 2. Exploratory Data Analysis

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='NDVI', y='LST_day', data=df, alpha=0.5, hue='ambient_temp')
plt.title('LST Day vs NDVI')
plt.show()
```

## 3. Modeling

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor

features = ['ambient_temp', 'humidity', 'wind_speed', 'NDVI', 'NDBI', 'albedo']
X = df[features].fillna(0)
y = df['LST_day']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

rf = RandomForestRegressor(n_estimators=50)
rf.fit(X_train, y_train)
print("R2 Score:", rf.score(X_test, y_test))
```
